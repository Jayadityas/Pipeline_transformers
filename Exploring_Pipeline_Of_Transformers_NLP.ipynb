```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier(["I am very excited today.","I am going to bid farewell."])
```

```python
classifier = pipeline('zero-shot-classification')
classifier("Our trip to america was very productive.",
           candidate_labels = ['lifestyle','education','healthcare'],)
```

```python
generator = pipeline('text-generation')
generator("On this auspicious occasion of republic day, we take pledge that",
          max_length = 100,
          num_return_sequences = 2,)
```

```python
unmasker = pipeline('fill-mask')
unmasker("This weekend the <mask> of US will take pledge as the president of US.", top_k = 3)
```

```python
ner = pipeline('ner', grouped_entities = True)
ner("John and Jessy are students of Michigan University and they belong to brooklyn.")
```

```python
quean = pipeline('question-answering')
quean(
    question = "Where do I work?",
    context = "My boss scolded me for being the most lethargic employee at Google.",
)
```

```python
sumamrizer = pipeline('summarization')
sumamrizer(""" It was during World War 2, when Einstein proposed his model for developing nuclear bomb. But Germany did not took him seriously, so he flew to
America. When he proposed his model before the US, he got a lot of recognition. But he was unaware of the fact that soon US would develop a nuclear bomb and drop it over Hiroshima
and Nagasaki in Japan. It was when Einstein realized that he unknowingly made a big mistake.""")
```

```python
translator = pipeline('translation',model='Helsinki-NLP/opus-mt-fr-en')
translator("Le modèle d’IA le plus puissant sera créé par moi.")
```
